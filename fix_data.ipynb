```python
def date_to_str(date):
    return date.strftime("%d-%m-%Y")
def rewrite_data(data: dict, date):
    house_csv = f"data/{date}_he.csv"
    bank_csv = f"data/{date}_bank.csv"
    data['he'].to_csv(house_csv, index = False)
    data['bank'].to_csv(bank_csv, index = False)
    
def add_day_col(date,  types = types_data):
    prev_data = load_day_data(date)
    data = prev_data.copy()
    date = date_to_str(date)
    for type in types:
        data[type] = data[type].assign(day = date)
    rewrite_data(data, date)
    return data
# load_day_data(dates[0])['bank']
[add_day_col(date) for date in dates]
# add_day_col(dates[0])
```
